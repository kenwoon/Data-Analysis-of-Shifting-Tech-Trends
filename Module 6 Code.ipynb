```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('job-postings.xlsx')
df.sort_values('Number of Job Postings', ascending=False).plot(kind='bar', x='Location', y='Number of Job Postings', figsize=(15, 5))

plt.title('Number of Job Postings by Location')
plt.xlabel('Job Postings')
plt.ylabel('Locations')
plt.show()
```

```python
df = pd.read_excel('popular-languages.xlsx')
df['Average Annual Salary'] = df['Average Annual Salary'].str.replace(',', '').str.replace('$', '').astype(int)
df.sort_values('Average Annual Salary', ascending=False).plot(kind='bar', x='Language', y='Average Annual Salary', figsize=(15, 5), legend=False)

plt.title('Average Annual Salary by Language Used')
plt.xlabel('Languages')
plt.ylabel('Salary in USD')
plt.show()
```
